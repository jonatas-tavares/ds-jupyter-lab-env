```python
import pandas as pd
```

```python
 df = pd.DataFrame({'A': range(1, 6),
                    'B': range(10, 0, -2),
                    'C': range(10, 5, -1)})
 print(df.query('A > B'))
 print(df.query('B == C'))
 print(df[df.A > df.B])
```

```python
s1 = pd.Series(['a', 'b'])
s2 = pd.Series(['c', 'd'])
df1 = pd.concat([s1, s2], axis=0, ignore_index=True)
df2 = pd.concat([s1, s2], axis=0)
print(df1)
print(df2)
df3 = df2.reset_index(drop=True)
print(df3)
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                     'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                     'value': [5, 6, 7, 8]})
print(df1.merge(
                df2, left_on='lkey', right_on='rkey', suffixes=(
                                                                '_left','_right'
                                                                )
                )
)
df3 = (df1.merge(
                df2, left_on='lkey', right_on='rkey', suffixes=(
                                                                '_left','_right'
                                                                )
                )
)
df3
```

```python
 df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                    'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
 other = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                       'B': ['B0', 'B1', 'B2']})
 df.join(other, lsuffix='_caller', rsuffix='_other')
```
